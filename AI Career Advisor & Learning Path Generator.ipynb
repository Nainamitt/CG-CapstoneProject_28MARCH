```python
from groq import Groq
import os
from dotenv import load_dotenv

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

MODEL_NAME = "llama-3.3-70b-versatile"
```

```python
def create_state(user_input):
    return {
        "input": user_input,
        "profile": None,
        "roles": None,
        "skill_gaps": None,
        "learning_path": None
    }
```

```python
def profile_analyzer(state):
    prompt = f"""
    Analyze the student profile.

    Extract:
    - skills
    - goal

    Input: {state['input']}
    """

    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    state["profile"] = response.choices[0].message.content.strip()
    return state
```

```python
def career_agent(state):
    prompt = f"""
    Based on this profile, suggest 3 suitable career roles.

    Profile:
    {state['profile']}
    """

    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    state["roles"] = response.choices[0].message.content.strip()
    return state
```

```python
def skill_gap_agent(state):
    prompt = f"""
    Identify missing skills required for these roles.

    Roles:
    {state['roles']}
    """

    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    state["skill_gaps"] = response.choices[0].message.content.strip()
    return state
```

```python
def learning_path_agent(state):
    prompt = f"""
    Create a step-by-step learning roadmap.

    Skill gaps:
    {state['skill_gaps']}
    """

    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    state["learning_path"] = response.choices[0].message.content.strip()
    return state
```

```python
def career_advisor_system(user_input):
    state = create_state(user_input)

    state = profile_analyzer(state)
    state = career_agent(state)
    state = skill_gap_agent(state)
    state = learning_path_agent(state)

    return state
```

```python
inputs = [
    "I know Python basics and statistics, and I want to become a Data Scientist",
    "I am good at HTML, CSS, and JavaScript, and I want to become a Software Developer"
]

for inp in inputs:
    print("\n==============================")
    print("User Input:", inp)
    
    result = career_advisor_system(inp)
    
    print("\n📊 Profile Analysis:\n", result["profile"])
    print("\n💼 Suggested Roles:\n", result["roles"])
    print("\n📉 Skill Gaps:\n", result["skill_gaps"])
    print("\n📚 Learning Path:\n", result["learning_path"])
```
